```python
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base-960h")
model.eval()
```

```python
def augment_waveforms(waveform, sr):
    augmented = [waveform]
    
    # Add noise
    noise = waveform + 0.005 * np.random.randn(len(waveform))
    augmented.append(noise)

    # Time stretch
    try:
        stretched = librosa.effects.time_stretch(waveform, rate=0.9)
        stretched = librosa.util.fix_length(stretched, len(waveform))
        augmented.append(stretched)
    except:
        pass

    # Time shift
    shifted = np.roll(waveform, int(sr * 0.1))
    augmented.append(shifted)

    return augmented
```

```python
def get_wav2vec2_embedding(waveform):
    inputs = processor(waveform, sampling_rate=16000, return_tensors="pt", padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return torch.mean(outputs.last_hidden_state, dim=1).squeeze().numpy()
```

```python
from tqdm import tqdm
import glob
import os
import librosa

emotions = {
    '01': 'neutral', '02': 'calm', '03': 'happy', '04': 'sad',
    '05': 'angry', '06': 'fearful', '07': 'disgust', '08': 'surprised'
}
observed_emotions = list(emotions.values())

X_fem, y_fem = [], []
X_male, y_male = [], []

files = glob.glob('C:\\Users\\Dell\\Downloads\\emotion_det\\common\\*.wav')
for file in tqdm(files, desc="Processing audio files", unit="file"):
    name_parts = os.path.basename(file).split("-")
    if len(name_parts) < 3:
        continue

    emo = emotions.get(name_parts[2])
    if emo not in observed_emotions:
        continue

    waveform, sr = librosa.load(file, sr=16000)
    gender = get_gender(file)
    gender_val = 1 if gender == 'female' else 0

    for aug_wave in augment_waveforms(waveform, sr):
        emb = get_wav2vec2_embedding(aug_wave)
        full_input = np.append(emb, gender_val)
        if gender == 'female':
            X_fem.append(full_input)
            y_fem.append(emo)
        else:
            X_male.append(full_input)
            y_male.append(emo)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# 1. Verify class distribution
print("Emotion distribution:")
print(pd.Series(y_male).value_counts(normalize=True))

# 2. Encode labels
le1 = LabelEncoder()
y_encoded1 = le1.fit_transform(y_male)
y_categorical1 = to_categorical(y_encoded1)

# 3. Prepare features
X1 = np.array(X_male)

# 4. Split data
x_train1, x_test1, y_train1, y_test1 = train_test_split(
    X1, y_categorical1, test_size=0.1, random_state=9
)

# 5. Feature scaling (CRITICAL)
scaler1 = StandardScaler()
x_train_scaled1 = scaler1.fit_transform(x_train1)
x_test_scaled1 = scaler1.transform(x_test1)

# 6. Build enhanced model
model1 = Sequential([
    Dense(512, activation='relu', input_shape=(x_train_scaled1.shape[1],)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(y_categorical1.shape[1], activation='softmax')
])

model1.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=0.0001),  # Lower LR
    metrics=['accuracy']
)
model1.summary()

# 7. Train with callbacks
history1 = model1.fit(
    x_train_scaled1, y_train1,
    epochs=200,  # More epochs
    batch_size=64,
    validation_data=(x_test_scaled1, y_test1),
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)
    ]
)

# 8. Evaluate and visualize
plt.plot(history1.history['accuracy'], label='Train')
plt.plot(history1.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()


```
